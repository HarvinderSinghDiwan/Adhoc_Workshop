```
#Data Visualisation
import  matplotlib.pyplot as plt
```

```
x=[10,20,25,30,45]
y=[1,3,5,7,8]
plt.xlabel("Distance")
plt.ylabel("Time")
plt.grid(color="blue")
plt.plot(x,y,label="railway track")
plt.scatter(x,y,label="Railway Station",marker="^",color="Red")
plt.bar(x,y,label="Railway Station Height",color="Violet")
plt.legend()
plt.show()
```

```
# CRICKET DATA VISUALISATION
players=["Dhoni","Virat","Harvinder"]
scores=[100,219,315]
plt.xlabel("Scores")
plt.ylabel("Players")
#plt.bar(players,scores,label="2020 score of some players")
plt.pie(scores,labels=players,explode=(0.2,0,0),autopct="%1.1ff%%")
plt.grid(color="blue")
plt.legend()
plt.show()
```
